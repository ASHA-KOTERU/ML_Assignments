```python
module = hub.Module("https://tfhub.dev/google/imagenet/inception_v3/feature_vector/1")
height, width = hub.get_expected_image_size(module)
```

```python
y_train=pd.read_csv("./data/labels.csv",delimiter=",")
y_true=pd.get_dummies(y_train,columns=['breed'])
#y_true.drop(['id'],axis=1,inplace=True)
y_true.columns = range(121)
```

```python

y_true.drop([0],axis=1,inplace=True)
y_true.head()

```

```python
y_true=y_true[:10000]
y_true = y_true.values
```

```python
y_true.shape
```

```python
type(y_true)
```

```python

batch_size=32
```

```python
X= tf.placeholder(tf.float32,[batch_size, 299,299,3])
y= tf.placeholder(tf.float32,[batch_size,120])
```

```python

# features from pretrained model
features=module(X)

#applying fully connected layer and softmax
logits=tf.layers.dense(inputs=features, units=120,activation=tf.nn.relu)
#prob=tf.nn.softmax(logits)

#calaculating loss
loss_op = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=logits, labels=y))

#
optimizer = tf.train.AdamOptimizer(learning_rate=0.0001)
train_op = optimizer.minimize(loss_op)


# Evaluate model

correct_pred = tf.equal(tf.argmax(logits, 1), tf.argmax(y, 1))
accuracy = tf.reduce_mean(tf.cast(correct_pred, tf.float32))
# Initialize the variables (i.e. assign their default value)
init = tf.global_variables_initializer()

```

```python
batch_size=32
```

```python

def get_next_batch(X,Y,s,batch_size):
    e=s+batch_size  
    return X[s:e],Y[s:e]
    
    
```

```python
with tf.Session() as sess:
    s=0
    sess.run(init)
    for step in range(309):   
        batch_x,batch_y=get_next_batch(images,y_true,s,batch_size)
        _,loss,acc=sess.run([train_op,loss_op,accuracy], feed_dict={X:batch_x,y:batch_y})
        print loss,acc
        s=s+batch_size
    print "optimization done"
    
```
